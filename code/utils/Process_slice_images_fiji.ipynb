# This code will preprocess your slice imaging data

Once you have run this code, you can import the results to Fiji to photobleach correct

You do NOT have to run both trimming and spatial downsampling. *The saved output will tell you if you performed spatial downsampling or trimming*

If all you care about is cell body activity, spatial downsampling is perfectly fine. Yet, if you need finer precision recordings (like dendrites), I would probably not downsample your data.

Downsampling your data preserves the structure of your image, but saves a lot of space on memory.

If you need full file, you should switch to a 64GB RAM machine at the very minimum.

The issue of memory propagates as you move to CaImAn. Smaller file sizes will run more smoothly and quickly. There is a tradeoff between how precise you need your data to be and how long or computationally expensive the analysis becomes.
